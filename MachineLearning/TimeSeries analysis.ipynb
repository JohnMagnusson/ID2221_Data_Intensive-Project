```python
### reading data 

data_loc = "../PreProcessing/data_cleaned/part-00000"

df = pd.DataFrame(pd.read_csv(data_loc, header=None,
                  names=[
                      'conversionSymbol',
                      'volumeto',
                      'high',
                      'low',
                      'time',
                      'volumefrom',
                      'close',
                      'open',
                      'conversionType'
                  ]
                  ))

print(df.head())
print(df.dtypes)
```

```python
### table to have a prior visualization of the data and its behavior

plt.figure(figsize = (18,9))
plt.plot(range(df.shape[0]),(df['low']+df['high'])/2.0)
plt.xticks(range(0,df.shape[0],500),df['time'].loc[::500],rotation=45)
plt.xlabel('time',fontsize=18)
plt.ylabel('Mid Price',fontsize=18)
plt.show()
```

```python
## splitting data into training / testing / validaiton
training_ratio = int(2/3 * df.shape[0])
training_df = df[:training_ratio]
testing_df = df[training_ratio:df.shape[0]-10]
validation_df = df[df.shape[0]-10:]
```
